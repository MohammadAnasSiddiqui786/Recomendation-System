```python
user_df3=df2.pivot_table(index='UserId',columns='BookTitle',values='BookRating').reset_index(drop=True)
user_df3
```

```python
user_df3.index=df2.UserId.unique()
user_df3
```

```python
user_df3.fillna(0, inplace=True)
user_df3
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_df3.values,metric='cosine')
user_sim
```

```python
user_sim_df3 = pd.DataFrame(user_sim)
```

```python
user_sim_df3.index = df2.UserId.unique()
user_sim_df3.columns = df2.UserId.unique()
```
